```python
def FFT(x,y,T):
    N=len(y)
    n=N/2
    df=1/T
    PSD=np.zeros(n)
    C=np.zeros(n)
    S=np.zeros(n)
    for i in range (0,n,1):
        for j in range (0,N,1):
            if i==0:
                C[i] += y[j]
            else:
                C[i] += np.cos(2*np.pi*(i/T)*x[j])*y[j]
                S[i] += np.sin(2*np.pi*(i/T)*x[j])*y[j]     
        if i==0:
            C[i]=C[i]/N
            PSD[i]= C[i]**2/df
            print "i=",i,C[i],PSD[i]
        else:
            C[i]= 2.0/N *C[i]
            S[i]= 2.0/N *S[i]
            PSD[i]= 0.5*((C[i]**2+S[i]**2)/df)
            print "i=",i,C[i],S[i],PSD[i]
    f=np.arange(0,n*df,df)
    pl.figure () 
    pl.plot(f,PSD)
    pl.xlabel('Hz')
    pl.ylabel('PSD')
    pl.show()
    return PSD,f
```
